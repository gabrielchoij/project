## 2-4 계산하기

```python
# 데이터가 정리되지 않은 상태에서 describe 함수를 통해 데이터를 이해하는 것은 위험합니다.
data.describe()
```

```python
# 섹터별 종가 평균
data.groupby('sector')['close_price'].mean()
```

```python
# 섹터 평균 열 만들기
data["sector_mean"] = data.groupby(['sector'])['close_price'].transform('mean')

# 섹터 티커 개수
data["sector_count"] = data.groupby(['sector'])['ticker'].transform('nunique')

data[["ticker", "sector", "sector_count", "close_price", "sector_mean"]]
```

```python
# 람다와 qcut 함수에 관한 자세한 내용은 해당 유튜브 강의 참고해 주세요.
data['price_s_level'] = data.groupby('sector')['close_price'].\
                        transform(lambda x: pd.qcut(x, q=3, labels=["L", "M", "H"]))

data[["ticker", "sector", "sector_count", "sector_mean", "close_price", "price_s_level"]]
```

```python
data[(data["sector"] == "Technology") & (data["price_s_level"] == "H")]\
                          [["ticker", "industry", "sector_mean", "close_price"]]
```
